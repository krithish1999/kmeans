```octave
initial_centroids = [3 3; 6 2; 8 5]
```

```octave
function idx = findClosestCentroid(X,centroids) 
K = size(centroids,1)
idx=zeros(size(X,1),1)
for i = 1:size(X,1)
    min = Inf;
    for j = 1:K
        diff = sum((X(i,:) - centroids(j,:)).^2);
        if min>diff
                min =  diff;
                idx(i) = j;
        end
    end
end

end
```

```octave
size(X,1)
```

```octave
idx = findClosestCentroid(X,initial_centroids) 
```

```octave
idx
```

```octave
function centroids = computerCentroids(X,idx,K)
[m,n] = size(X);
centroids = zeros(K,n);
ns = zeros(K,1);
sums = zeros(K,n);
for i = 1:size(idx,1)
    z = idx(i);
    ns(z)=ns(z)+1;
    sums(z,:)=sums(z,:)+X(i,:);
end
centroids = sums./ns;
end
```

```octave
centroids = computerCentroids(X,idx,K)
```

```octave
X, idx
```

```octave
idx = findClosestCentroid(X,centroids) 
```

```octave
centroids = computerCentroids(X,idx,K)
```

```octave
function plotDataPoints(X, idx, K)
palette = hsv(K + 1);
colors = palette(idx, :);
scatter(X(:,1), X(:,2), 100, colors);
end
```

```octave
plotDataPoints(X,idx,K)
```
